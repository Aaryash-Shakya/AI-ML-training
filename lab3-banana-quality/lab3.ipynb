```python
for column_name in [
    "Softness",
    "HarvestTime",
    "Quality",
    "Weight",
    "Size",
    "Sweetness",
    "Ripeness",
    "Acidity",
]:
    # Assuming df is your DataFrame and 'column_name' is the column for which you want to check outliers
    Q1 = train_data[column_name].quantile(0.25)
    Q3 = train_data[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define the boundary for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Get the number of outliers
    outliers = train_data[
        (train_data[column_name] < lower_bound)
        | (train_data[column_name] > upper_bound)
    ]
    num_outliers = outliers.shape[0]

    print(f"There are {num_outliers} outliers in {column_name}")
    # replace with nan for outliers
    train_data[f"Adjusted_{column_name}"] = np.where(
        (train_data[column_name] < lower_bound)
        | (train_data[column_name] > upper_bound),
        np.nan,
        train_data[column_name],
    )
    # Calculate the mean of the non-outlier values
    mean = train_data[f"Adjusted_{column_name}"].median()

    # Replace the NaNs in the 'Adjusted' column with the mean
    train_data[f"Adjusted_{column_name}"].fillna(mean, inplace=True)
    train_data = train_data.drop(columns=column_name, axis=1)
```

```python
for column_name in [
    "Adjusted_Softness",
    "Adjusted_HarvestTime",
    "Adjusted_Quality",
    "Adjusted_Weight",
    "Adjusted_Size",
    "Adjusted_Sweetness",
    "Adjusted_Ripeness",
    "Adjusted_Acidity",
]:
    # Assuming df is your DataFrame and 'column_name' is the column for which you want to check outliers
    Q1 = train_data[column_name].quantile(0.25)
    Q3 = train_data[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define the boundary for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Get the number of outliers
    outliers = train_data[
        (train_data[column_name] < lower_bound)
        | (train_data[column_name] > upper_bound)
    ]
    num_outliers = outliers.shape[0]

    print(f"There are {num_outliers} outliers in {column_name}")

    # replace with nan for outliers
    train_data[f"New_{column_name[9:]}"] = np.where(
        (train_data[column_name] < lower_bound)
        | (train_data[column_name] > upper_bound),
        np.nan,
        train_data[column_name],
    )
    # Calculate the mean of the non-outlier values
    mean = train_data[f"New_{column_name[9:]}"].median()

    # Replace the NaNs in the 'Adjusted' column with the mean
    train_data[f"New_{column_name[9:]}"].fillna(mean, inplace=True)
    train_data = train_data.drop(columns=column_name, axis=1)
```

```python
for column_name in [
    "New_Softness",
    "New_HarvestTime",
    "New_Quality",
    "New_Weight",
    "New_Size",
    "New_Sweetness",
    "New_Ripeness",
    "New_Acidity",
]:
    # Assuming df is your DataFrame and 'column_name' is the column for which you want to check outliers
    Q1 = train_data[column_name].quantile(0.25)
    Q3 = train_data[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define the boundary for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Get the number of outliers
    outliers = train_data[
        (train_data[column_name] < lower_bound)
        | (train_data[column_name] > upper_bound)
    ]
    num_outliers = outliers.shape[0]

    print(f"There are {num_outliers} outliers in {column_name}")
```

```python
features = ["New_Weight", "New_Size", "New_Sweetness", "New_Ripeness", "New_Acidity"]
label = ["New_Quality"]
from sklearn.model_selection import train_test_split

# Assuming you have your featues and target variables in the following format
X_train, X_test, Y_train, Y_test = train_test_split(
    train_data[features], train_data[label], test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
Y_pred = logistic_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy*100}%")
```
